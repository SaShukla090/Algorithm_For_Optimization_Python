## Chapter3: Bracketing

#### Algorithm3.1

```python
def bracket_minimum(f: object, x: float= 0.0, s: float=1e-2, k: float=2.0):
    """
    Brackets the minimum of a unimodal function.
    This function attempts to find an interval (a, c) that contains a local minimum of the given unimodal function f.
    It starts from an initial point x and step size s, and iteratively expands the interval by a factor of k until
    the function value at the new point is greater than the function value at the previous point.
    Parameters:
    f (object): The unimodal function to minimize.
    x (float, optional): The initial point to start the search. Default is 0.0.
    s (float, optional): The initial step size for expanding the interval. Default is 1e-2.
    k (float, optional): The factor by which to expand the interval in each iteration. Default is 2.0.
    Returns:
    tuple: A tuple (a, c) representing the interval that contains the local minimum.
    
    """

    a,Ya = x, f(x)
    b,Yb = x + s, f(x + s)
    if Yb > Ya:
        a,b = b,a
        Ya,Yb = Yb, Ya
        s = -s
    
    while True:
        c, Yc = b+s, f(b + s)
        if Yc > Yb:
            return (a,c) if a<c else (c,a)
        a,Ya,b,Yb = b, Yb, c, Yc
        s *= k
```

```python
import numpy as np

import matplotlib.pyplot as plt

# Define a unimodal function for testing
def test_function(x):
    return (x + 2) ** 2 + 3

# Generate test case
initial_point = 0.0
step_size = 1e-2
expansion_factor = 2.0

# Call the bracket_minimum function
interval = bracket_minimum(test_function, initial_point, step_size, expansion_factor)
print(f"Bracketed interval: {interval}")

# Plot the function and the bracketed interval
x = np.linspace(-10, 10, 400)
y = test_function(x)
plt.plot(x, y, label='Test Function')
plt.axvline(interval[0], color='r', linestyle='--', label='Interval Start')
plt.axvline(interval[1], color='g', linestyle='--', label='Interval End')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Bracketed Interval for Minimum')
plt.show()
```
